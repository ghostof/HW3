```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,1,0.1)
print(x)
```

```python
rnd = np.random.rand(1,10)
print(rnd)
```

```python
f1 = lambda x: 20+(7.5*x)+(3*x)+(13*x)
print(f1(x))
f2 = lambda x: 20+(7.5*x)+(3*x**2)+(13*x**3)
print(f2(x))
```

```python
fig,ax = plt.subplots()
ax.plot(x,f1(x))
plt.show()
```

```python
fig,ax = plt.subplots()
ax.plot(x,f2(x))
plt.show()
```
